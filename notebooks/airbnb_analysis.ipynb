```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/AB_NYC_2019.csv")

# Quick look at dataset
print(df.head(), "\n")
print(df.info(), "\n")
print(df.describe(), "\n")

# Listings per borough
plt.figure(figsize=(8,5))
sns.countplot(x="neighbourhood_group", data=df, order=df['neighbourhood_group'].value_counts().index)
plt.title("Airbnb Listings by Borough")
plt.show()

# Price distribution (remove outliers > 500)
plt.figure(figsize=(10,6))
sns.boxplot(x="neighbourhood_group", y="price", data=df[df["price"] < 500])
plt.title("Price Distribution by Borough (Filtered)")
plt.show()

# Top 10 neighborhoods
top_neighborhoods = df['neighbourhood'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_neighborhoods.values, y=top_neighborhoods.index, palette="viridis")
plt.title("Top 10 Neighborhoods by Number of Listings")
plt.xlabel("Number of Listings")
plt.show()

# Room type distribution
room_counts = df['room_type'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=room_counts.index, y=room_counts.values, palette="muted")
plt.title("Room Type Distribution")
plt.ylabel("Number of Listings")
plt.show()
```

# 📊 Insights from Airbnb NYC Data

1. **Listings are concentrated in Manhattan and Brooklyn** – these two boroughs dominate the market.  
2. **Price distribution shows wide variation**, with Manhattan having the highest median prices.  
3. **Top neighborhoods** such as Williamsburg, Harlem, and Bedford-Stuyvesant lead in the number of listings.  
4. **Entire home/apartment** is the most common room type, but private rooms also have strong representation.  
5. Filtering out extreme outliers (listings > $500) gives a clearer view of price patterns across boroughs.  

✅ These insights can help investors, hosts, and policymakers better understand Airbnb dynamics in NYC.
